```python
import numpy as np
import os
import cv2
from skimage import io
from skimage import transform
import matplotlib.pyplot as plt
```

```python
def affine(img, offset):
    #获取图像大小
    rows, cols = img.shape[:2]

    #设置图像仿射变换矩阵
    pos1 = np.float32([[0,0], [400,250], [100,250]])
    pos2 = np.float32([[0,0], [400,250], [100+offset,250]])
    M = cv2.getAffineTransform(pos1, pos2)

    #图像仿射变换
    result = cv2.warpAffine(img, M, (cols, rows))

    return result
```

```python
dataset_path = 'd:/datasets/jz20/'
path_list = os.listdir(dataset_path)
```

```python
for path in path_list:
    for i in range(0,11):
        print( dataset_path + path + '/' + str(i) +  '.jpg')
        file_name2 = dataset_path + path + '/' + str(i) +  '.jpg'
        img_src=io.imread(file_name2)
        img_dst_1=affine(img_src, 20)        
        cv2.imencode('.jpg', img_dst_1[:,:,::-1])[1].tofile(dataset_path + path + '/' + str(i) + '1.jpg')
        img_dst_2=affine(img_src, -20)        
        cv2.imencode('.jpg', img_dst_2[:,:,::-1])[1].tofile(dataset_path + path + '/' + str(i) + '2.jpg')
```
