```python
import numpy as np
import os
import cv2
from skimage import io
from skimage import transform
import matplotlib.pyplot as plt
```

```python
def Linear_dodge(img_1, img_2):
    img = img_1 + img_2
    mask_2 = img  > 1
    img = img * (1-mask_2) + mask_2
    return img
```

```python
dataset_path = 'd:/datasets/jz20/'
path_list = os.listdir(dataset_path)
```

```python
for path in path_list:
    img_1=io.imread(dataset_path + path + '/0.jpg')
    img_1=transform.resize(img_1, (400, 250))
    img_1 = img_1/255.0
    
    for i in range(1,11):
        file_name2='mask/' + str(i) +  '.jpg'
        img_2=io.imread(file_name2)
        img_2=transform.resize(img_2, (400, 250))
        img_2 = img_2/255.0

        img = Linear_dodge(img_2, img_1)
        io.imsave(dataset_path + path + '/' + str(i) + '.jpg',(img*255))
```
